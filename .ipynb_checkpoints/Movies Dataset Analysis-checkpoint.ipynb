```python
# As we can see that columns Budget & Gross are float so we need to convert the data type to be integers 

df['budget'] = df['budget'].astype('int64')
df['gross'] = df['gross'].astype('int64')
```

```python
# now we can drop the year column as we don't need it anymore 
df.drop(['year'], axis = 1, inplace = True)
```

My hypothesis here is that budget is correlated with the gross revenue for each movie so let's zoom in and focus on budget VS. gross columns using a scatter plot 

```python
# plot a scatter plot for budget and gross 
plt.scatter(x = df['budget'], y = df['gross'], alpha = 0.5, color = 'blue')
plt.title('Budget VS. Gross Revenue')
plt.xlabel('Budget')
plt.ylabel('Gross Revenue')
plt.show()
```

```python
# Use reg plot to determine the correlation between budget & gross revenue 
sns.regplot(x = 'budget', y = 'gross', data = df, scatter_kws = {'color':'blue'}, line_kws = {'color':'black'})
```

As we can see there are other non numeric features like company or country or director that could affect the gross revenue of each movie and we couldn't check these correlations before with the above heatmap since these features are not numeric so we will convert the unique values of these features from strings or objects to numeric values to be able to check if there is any feature that affect te gross revenue. 

```python
# First we will take a copy of our dataframe and work with the copy to keep the original dataframe as it's in case we need to get back to it
numeric_df = df.copy()
```

```python
numeric_df.head()
```

```python
# we will use for loop to change all of the features with object data type to be category 

for col in numeric_df.columns:
    if (numeric_df[col].dtype == 'object'):
        numeric_df[col] = numeric_df[col].astype('category')
        numeric_df[col] = numeric_df[col].cat.codes
```

```python
numeric_df.head(20)
```

Now we can see that all the non numeric features now cointain numeric values (unique number for each value) and now we can check the correlations between these features and plot them. 

```python
numeric_df.dtypes
```

```python
# Check the correlation matrix 
correlation_mat = numeric_df.corr()

# This piece of code is to order the features based on the correlation score in descending order 
corr_pairs = correlation_mat.unstack()
corr_pairs
```

```python
# We can sort the values to be easier for us to read 
sorted_pairs = corr_pairs.sort_values()
sorted_pairs
```

```python
# Check the high correlation scores 

high_corr = sorted_pairs[(sorted_pairs) > 0.5]
high_corr
```

From the above we can conclude that our hypothesis is not significant as we can see that the other features like company, country, genre and ext are not correlated with gross revenue. But we also discovered that votes features is highly correlated with gross revenue.
